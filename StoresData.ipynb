```python
# Necessary imports for web-scraping
import requests 
from bs4 import BeautifulSoup
```

```python
# Beginning the web-scraping process
# First we are creating a list of links 
url_list = storesData['Link'].tolist()
print(url_list)

#Creating an empty list to store the title of each product and one for the price of each product
titles = []
prices = []

# Now for every url in the url_list we will grab its html and parse it
for url in url_list:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract title (update selectors based on actual webpage)
    title_tag = soup.select_one('h1')  
    title = title_tag.text.strip() if title_tag else 'N/A'

    # Extract price (update selectors based on actual webpage)
    price_tag = soup.select_one('.aditem-main--middle--price, .price') 
    price = price_tag.text.strip() if price_tag else 'N/A'

    # Append data to lists
    titles.append(title)
    prices.append(price)

# Create pandas DataFrame for structured view
priceTable = pd.DataFrame({
    'Title': titles,
    'Price': prices,
    'URL': url_list
})
```
